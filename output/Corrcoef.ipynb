```python
country = pd.read_csv('../input/climate-change-earth-surface-temperature-data/GlobalLandTemperaturesByCountry.csv')
country['dt'] = pd.to_datetime(country['dt'])
country['dt']
mask = (country['dt'] > '1980-01-01')
country = country.loc[mask]




countriesWithTemp = {}
for nameOfCountry in country.Country.unique().tolist():
    countriesWithTemp[nameOfCountry] = country.loc[country['Country'] == nameOfCountry].AverageTemperature.mean()
    
# for tempNum in countriesWithTemp.values():
#     print(tempNum)

fig3 = plt.figure()
ax1 = fig3.add_subplot(111)
ax1.scatter(list(countriesWithTemp.keys()), list(countriesWithTemp.values()), s=10, c='b', marker="o", label='Countries with Average Temperature')
plt.legend(loc='upper left');
plt.show()

countriesWithTemp
```

```python


gdp = pd.read_csv('../input/gross-domestic-product/GDP.csv')

gdp = gdp.iloc[4:, 3:5]
gdp.rename(columns={'Unnamed: 3': 'Country', 'Unnamed: 4': 'GDP'}, inplace=True)

gdp = gdp.dropna()

for i in range(len(gdp['GDP'])):
    try:
        gdp['GDP'].iloc[i] = int(gdp['GDP'].iloc[i][1:-2].replace(',', ''))
    except:
        continue
gdp = gdp[gdp.GDP != ' .. ']

# def convertFunc(x):
#     try:
#         return int(x[1:-2].replace(',', ''))
#     except:
#         return

fig4 = plt.figure()
ax1 = fig4.add_subplot(111)
ax1.scatter(list(gdp['Country']), list(gdp['GDP']), s=10, c='b', marker=".", label='Countries to their Gross Domestic Product')
plt.legend(loc='upper left');
plt.show()



# len(gdp['GDP'])
int(gdp.loc[gdp['Country'] == 'United States'].GDP)
```
